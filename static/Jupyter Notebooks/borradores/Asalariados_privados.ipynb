```python
"|IMPORT PACKAGES|"
import pandas             as pd
import datetime
from   bokeh.plotting     import show, figure, output_file, save
from   bokeh.io           import show, output_notebook, curdoc, export_png
from   bokeh.models       import ColumnDataSource,LinearAxis, Range1d, NumeralTickFormatter, LabelSet, Label, BoxAnnotation, DatetimeTickFormatter, Legend, LegendItem
from   bokeh.models.tools import HoverTool
```

```python
"|IMPORT DATA|"
path = r'https://github.com/ncachanosky/research/blob/master/Economic%20Series/'
file = r'Resumen%20Estadistico%20-%20Argentina.xlsx?raw=true'
IO   = path + file

sheet = 'MONTHLY'

data_mensual = pd.read_excel(IO, sheet_name = sheet, skiprows=2,usecols="B,JB,JC",engine='openpyxl')
data=data_mensual.dropna()
data
```

```python
cds = ColumnDataSource(data)

#BUILD FIGURE
p = figure(title        = "EL HUB ECONÓMICO | ASALARIADOS SECTOR PRIVADO",
        
           plot_height  = 400,
           plot_width   = 700,
           x_axis_type = "datetime",
           y_axis_label = 'Asalariados privados (en millones)')

p.toolbar_location = 'right'
p.toolbar.autohide = True


#AXIS 1 (LEFT)
p1= p.line("MONTH", "Asalariados privados", color="cyan",width=2,legend_label='Asalariados privados',source=cds)  
p.legend.location     = 'top_left'
p.legend.orientation  = 'vertical'
p.legend.click_policy = 'mute'

show(p)

```
