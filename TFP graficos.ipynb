```python
import pandas             as pd
import datetime
from   bokeh.plotting     import show, figure, output_file, save
from   bokeh.io           import show, output_notebook, curdoc, export_png
from   bokeh.models.tools import HoverTool
from   bokeh.models       import ColumnDataSource,LinearAxis, Range1d, NumeralTickFormatter, LabelSet, Label, BoxAnnotation, DatetimeTickFormatter, Text, Span
from   bokeh.models       import Arrow, NormalHead, OpenHead, VeeHead
from   bokeh.transform    import dodge
from   datetime           import datetime as dt
```

```python
"|IMPORT DATA|"
path = r'https://github.com/ncachanosky/research/blob/master/Economic%20Series/'
file = r'Resumen%20Estadistico%20-%20Internacional.xlsx?raw=true'
IO   = path + file
sheet = 'DATA'

data = pd.read_excel(IO, sheet_name = sheet, usecols="A:C, S", skiprows=2, nrows=12318, engine='openpyxl') # Be patient...

data = data.dropna()

data = data.rename(columns={"TFP (CONSTANT 2007)":"TFP"})
```

```python
"|CHECK DATA|"
data.tail()
```

```python
data_arg = data.loc[data["COUNTRY"]=="Argentina",:]
```

```python
data_arg
```

```python
cds = ColumnDataSource(data_arg)

#BUILD FIGURE
p = figure(title = "EL HUB ECONÓMICO | PRODUCTIVIDAD TOTAL DE LOS FACTORES EN ARGENTINA",
           x_axis_label = "Año", 
           plot_height  = 400,
           plot_width   = 700)

p.toolbar_location = "right"
p.toolbar.autohide = True
p1 = p.line("YEAR", "TFP", color="blue", line_alpha=0.50, width=4, legend_label="TFP"    ,
            muted_alpha=0.20, source=cds)

p.add_tools(HoverTool(renderers=[p1], tooltips = [("TFP"    , "@TFP")], mode="vline"))

#anotaciones
convert_text = Label(x=1985, y=0.83, text_font_size='12px', text_color='red', text_align='center',
                 text='Convertibilidad')
convert_arrow = Arrow(end=VeeHead(line_color='red', line_width=2, line_alpha=0.5, size=10, fill_color='red', fill_alpha=0.75),
                  x_start=1985, x_end=1990, y_start=0.83, y_end=0.82,
                  line_color='red', line_alpha=0.5)
tequila_text = Label(x=1997, y=1.1, text_font_size='12px', text_color='red', text_align='center',
                 text='Crisis del Tequila')
tequila_arrow = Arrow(end=VeeHead(line_color='red', line_width=2, line_alpha=0.5, size=10, fill_color='red', fill_alpha=0.75),
                  x_start=1997, x_end=1997, y_start=1.1, y_end=1.06,
                  line_color='red', line_alpha=0.5)
devalua_text = Label(x=2008, y=0.835, text_font_size='12px', text_color='red', text_align='center',
                 text='Mega devaluacion')
devalua_arrow = Arrow(end=VeeHead(line_color='red', line_width=2, line_alpha=0.5, size=10, fill_color='red', fill_alpha=0.75),
                  x_start=2008, x_end=2002, y_start=0.85, y_end=0.864,
                  line_color='red', line_alpha=0.5)
c2008_text = Label(x=2008, y=1.15, text_font_size='12px', text_color='red', text_align='center',
                 text='Crisis Subprime')
c2008_arrow = Arrow(end=VeeHead(line_color='red', line_width=2, line_alpha=0.5, size=10, fill_color='red', fill_alpha=0.75),
                  x_start=2008, x_end=2008, y_start=1.15, y_end=1.089,
                  line_color='red', line_alpha=0.5)
estanf_text = Label(x=2016, y=1.1, text_font_size='12px', text_color='red', text_align='center',
                 text='Inicio de estanflacion')
estanf_arrow = Arrow(end=VeeHead(line_color='red', line_width=2, line_alpha=0.5, size=10, fill_color='red', fill_alpha=0.75),
                  x_start=2016, x_end=2011, y_start=1.1, y_end=1.095,
                  line_color='red', line_alpha=0.5)
inflatarg_text = Label(x=2012, y=0.95, text_font_size='12px', text_color='red', text_align='center',
                 text='Cambio de metas de inflacion')
inflatarg_arrow = Arrow(end=VeeHead(line_color='red', line_width=2, line_alpha=0.5, size=10, fill_color='red', fill_alpha=0.75),
                  x_start=2012, x_end=2017, y_start=0.965, y_end=1,
                  line_color='red', line_alpha=0.5)
rodrigazo_text = Label(x=1974, y=1.04, text_font_size='12px', text_color='red', text_align='center',
                 text='Rodrigazo')
rodrigazo_arrow = Arrow(end=VeeHead(line_color='red', line_width=2, line_alpha=0.5, size=10, fill_color='red', fill_alpha=0.75),
                  x_start=1974, x_end=1974, y_start=1.05, y_end=1.146,
                  line_color='red', line_alpha=0.5)
austral_text = Label(x=1980, y=0.9, text_font_size='12px', text_color='red', text_align='center',
                 text='Plan Austral')
austral_arrow = Arrow(end=VeeHead(line_color='red', line_width=2, line_alpha=0.5, size=10, fill_color='red', fill_alpha=0.75),
                  x_start=1980, x_end=1985, y_start=0.9, y_end=0.886,
                  line_color='red', line_alpha=0.5)
p.add_layout(austral_text)
p.add_layout(austral_arrow)
p.add_layout(rodrigazo_text)
p.add_layout(rodrigazo_arrow)
p.add_layout(inflatarg_text)
p.add_layout(inflatarg_arrow)
p.add_layout(estanf_text)
p.add_layout(estanf_arrow)
p.add_layout(c2008_text)
p.add_layout(c2008_arrow)
p.add_layout(devalua_text)
p.add_layout(devalua_arrow)
p.add_layout(convert_text)
p.add_layout(convert_arrow)
p.add_layout(tequila_text)
p.add_layout(tequila_arrow)

p.legend.location     = "top_left"
p.legend.orientation  = "horizontal"
p.legend.click_policy = "mute"
show(p)
```

```python
cdsarg = ColumnDataSource(data_arg)
cdschi = ColumnDataSource(data_chi)
cdscol = ColumnDataSource(data_col)
cdsbra = ColumnDataSource(data_bra)

#BUILD FIGURE
p = figure(title = "EL HUB ECONÓMICO | PRODUCTIVIDAD TOTAL DE LOS FACTORES EN ARGENTINA VS OTROS PAISES ",
           x_axis_label = "Año", 
           plot_height  = 400,
           plot_width   = 700)

p.toolbar_location = "right"
p.toolbar.autohide = True
p1 = p.line("YEAR", "TFP", color="blue", line_alpha=0.50, width=4, legend_label="TFP ARGENTINA"    ,
            muted_alpha=0.20, source=cdsarg)
p2 = p.line("YEAR", "TFP", color="red", line_alpha=0.50, width=4, legend_label="TFP CHILE"    ,
            muted_alpha=0.20, source=cdschi)
p3 = p.line("YEAR", "TFP", color="yellow", line_alpha=0.50, width=4, legend_label="TFP COLOMBIA"    ,
            muted_alpha=0.20, source=cdscol)
p4 = p.line("YEAR", "TFP", color="green", line_alpha=0.50, width=4, legend_label="TFP BRASIL"    ,
            muted_alpha=0.20, source=cdsbra)


p.add_tools(HoverTool(renderers=[p1,p2,p3,p4], tooltips = [("TFP"    , "@TFP")], mode="vline"))
p.legend.location     = "top_right"
p.legend.orientation  = "vertical"
p.legend.click_policy = "mute"
show(p)
```
