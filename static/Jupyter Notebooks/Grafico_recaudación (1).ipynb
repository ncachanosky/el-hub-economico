```python
"|IMPORT PACKAGES|"
import pandas             as pd
import datetime
from   bokeh.plotting     import show, figure, output_file, save
from   bokeh.io           import show, output_notebook, curdoc, export_png
from   bokeh.models       import ColumnDataSource,LinearAxis, Range1d, NumeralTickFormatter, LabelSet, Label, BoxAnnotation, DatetimeTickFormatter, Legend, LegendItem
from   bokeh.models.tools import HoverTool
```

```python
"|IMPORT DATA|"
path = r'https://github.com/ncachanosky/research/blob/master/Economic%20Series/'
file = r'Resumen%20Estadistico%20-%20Argentina.xlsx?raw=true'
IO   = path + file

sheet = 'REC TRIB'

data_mensual = pd.read_excel(IO, sheet_name = sheet, skiprows=4, engine='openpyxl')
```

```python
data= data_mensual.groupby(pd.Grouper(key='Anuales', freq='Y')).sum()
data
```

```python
data["G/T"]=data[" Ganancias"]/data[" Total recursos tributarios"]*100
data["IVA/T"]=data[" IVA"]/data[" Total recursos tributarios"]*100
data["TC/T"]=data["Total Combustibles"]/data[" Total recursos tributarios"]*100
data["BP/T"]=data[" Bienes Personales"]/data[" Total recursos tributarios"]*100
data["EX/T"]=data[" Derechos de Exportación"]/data[" Total recursos tributarios"]*100
data["IM/T"]=data[" Derechos de Importación y Otros"]/data[" Total recursos tributarios"]*100
data["SS/T"]=data[" Sistema Seguridad Social"]/data[" Total recursos tributarios"]*100
data["O/T"]=(100-(data["G/T"]+data["IVA/T"]+data["TC/T"]+data["BP/T"]+data["EX/T"]+data["IM/T"]+data["SS/T"]))



```

```python
cds = ColumnDataSource(data)

p = figure(title = "EL HUB ECONÓMICO | PARTICIPACIÓN DE IMPUESTOS EN RECAUDACIÓN",
           y_axis_label = "Participación por impuesto", 
           plot_height  = 600,
           plot_width   = 700,
          x_axis_type = "datetime")

p.toolbar_location = "right"
p.toolbar.autohide = True

colors=['black',     # 1
        'blue' ,     # 2
        'red'  ,     # 3
        'green',     # 4
        'slateblue', # 5
        'teal',      # 6
        'tan',       # 7
        'plum'  ]    # 8

p1 = p.varea_stack(["G/T","IVA/T","TC/T","BP/T","EX/T","IM/T","SS/T","O/T"], x='Anuales', color=colors, fill_alpha=0.5, muted_alpha=0.20, source=cds)
p2 = p.vline_stack(["G/T","IVA/T","TC/T","BP/T","EX/T","IM/T","SS/T","O/T"], x='Anuales', color=colors, line_alpha=1.0, muted_alpha=0.20, source=cds
                  
                  )



legends=["Ganancias",
        "IVA",
        "Combustibles",
        "Bienes Personales",
        "Derechos de Exportación",
        "Derechos de Importación",
        "Seguridad Social",
        "Otros impuestos"]
legend1 = Legend(items=[(legends[ 0], [p1[ 0]]),
                        (legends[ 1], [p1[ 1]]),
                        (legends[ 2], [p1[ 2]]),
                        (legends[ 3], [p1[ 3]]),
                       (legends[ 4], [p1[ 4]])], orientation="horizontal", click_policy='mute')

legend2 = Legend(items=[(legends[ 5], [p1[ 5]]),
                       (legends[ 6], [p1[ 6]]),
                       (legends[ 7], [p1[ 7]])], orientation="horizontal", click_policy='mute')

p.add_layout(legend1, "below")
p.add_layout(legend2, "below")



show(p)
```
