```python
"|IMPORT PACKAGES|"
import pandas             as pd
from   bokeh.plotting     import show, figure, output_file, save
from   bokeh.io           import show, output_notebook
from   bokeh.io           import curdoc
from   bokeh.models       import LinearAxis, Range1d
from   bokeh.models       import ColumnDataSource
from   bokeh.models.tools import HoverTool
from   bokeh.transform    import dodge
```

```python
"|IMPORT DATA|"
path = r'https://github.com/ncachanosky/research/blob/master/Economic%20Series/'
file = r'Resumen%20Estadistico%20-%20Argentina.xlsx?raw=true'
IO   = path + file

sheet = 'PBI'

data = pd.read_excel(IO, sheet_name = sheet, usecols="A:F", skiprows=10, nrows=27) # Be patient...
data = data.rename(columns={"%PBI (1993)":"1993"})
data = data.rename(columns={"%PBI (2004)":"2004"})
data = data.rename(columns={"2015 = 100":"Y_100"})
cds = ColumnDataSource(data)
```

```python
"|CHECK DATA|"
data
```

```python
"|BUILD PLOT|"
#BUILD FIGURE
p = figure(title        = "PBI real (1994 - 2019)",
           x_axis_label = "Año",
           y_axis_label = "PBI (tasa de crecimiento)",
           x_range      = (1994, 2019),
           y_range      = (-15, 15),
           plot_height  =  600,
           plot_width   = 1000)

p.toolbar_location = "above"
p.toolbar.autohide = True
p.add_tools(HoverTool(tooltips = [("Año", "@YEAR"), ("%PBI (1993)", "@1993"), ("%PBI (2004)", "@2004"), ("2005=100", "@Y_100")]))


#AXIS 1 (LEFT)
p.vbar(x=dodge('YEAR', -0.20, range=p.x_range), top='1993', color="red" , width=0.25, fill_alpha=0.50, legend_label="Base 1993", source=cds)
p.vbar(x=dodge('YEAR',  0.20, range=p.x_range), top='2004', color="blue", width=0.25, fill_alpha=0.50, legend_label="Base 2004", source=cds)

#AXIS 2 (RIGHT)
p.extra_y_ranges = {"y2": Range1d(start=0, end=120)}
p.add_layout(LinearAxis(y_range_name = "y2", axis_label = "2005 = 100"), 'right')
p.line  ('YEAR', 'Y_100', color="grey", y_range_name = "y2", source=cds, legend_label="2005 = 100")
p.circle('YEAR', 'Y_100', color="grey", y_range_name = "y2", source=cds, size=7)

#LEGEND
p.legend.location    ="top_left"
p.legend.orientation ="horizontal"
show(p)
```

```python
# CREATE HTML FILE

output_file(filename="PBI real.html", title="PBI real")
save(p)
```

```python
# CREATE JSON FILE
import json
import bokeh.embed

from bokeh.embed import json_item

j = json.dumps(json_item(p, "PBI real"))
with open("PBI Real.json", "w") as fp:
    json.dump(j, fp)
```
